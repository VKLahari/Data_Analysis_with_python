```python
df['MonthlyCharges'].fillna(df['MonthlyCharges'].mean(), inplace=True)
```

```python
df.drop_duplicates(inplace=True)
print(df['Churn'].unique())

# Clean data if necessary (e.g., strip spaces, capitalize, replace values)
df['Churn'] = df['Churn'].astype(str).str.strip()
df['Churn'] = df['Churn'].str.capitalize()
```

```python
df['Churn'] =df['Churn'].map({'Yes': 1, 'No': 0})
df.head(5)
```

```python
df['Churn'].value_counts()
```

```python
print(df['gender'].unique())
df['gender'] = df['gender'].astype(str).str.strip()
df['gender'] = df['gender'].str.capitalize()
```

```python
df['gender'] =df['gender'].map({'Female': 1, 'Male': 0})
df.head(5)
```

```python
print(df['gender'].isnull().sum())
```

```python
df['gender'].value_counts()
```

```python
print(df['Partner'].unique())
df['Partner'] = df['Partner'].astype(str).str.strip()
df['Partner'] = df['Partner'].str.capitalize()
```

```python
df['Partner']=df['Partner'].map({'Yes':1,'No':0})
df.head(5)
```

```python
df['Dependents'].unique()
```

```python
df['Dependents'] = df['Dependents'].astype(str).str.strip()
df['Dependents'] = df['Dependents'].str.capitalize()
```

```python
df['Dependents']=df['Dependents'].map({'Yes':1,'No':0})
df.head(5)
```

```python
print(df['PhoneService'].unique())
df['PhoneService'] = df['PhoneService'].astype(str).str.strip()
df['PhoneService'] = df['PhoneService'].str.capitalize()
```

```python
df['PhoneService']=df['PhoneService'].map({'Yes':1,'No':0})
df.head(5)
```

```python
print(df['PhoneService'].isnull().sum())
```

```python
print(df['PaperlessBilling'].unique())
df['PaperlessBilling'] = df['PaperlessBilling'].astype(str).str.strip()
df['PaperlessBilling'] = df['PaperlessBilling'].str.capitalize()
```

```python
print(df.columns)
```

```python
df.info()
```

```python
df.head(5)
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df['TotalCharges'].fillna(df['TotalCharges'].mean(), inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
df[['tenure','MonthlyCharges','TotalCharges']]=scaler.fit_transform(df[['tenure','MonthlyCharges','TotalCharges']])
```

```python
df.head(5)
```

**Exploratory Data Analysis**

```python
import matplotlib.pyplot as plt
contract_churn = df.groupby('Contract')['Churn'].mean()  # Calculate churn rate for each contract type

plt.figure(figsize=(8, 6))
contract_churn.plot(kind='bar', color='orange')
plt.title('Churn Rate by Contract Type')
plt.xlabel('Contract')
plt.ylabel('Churn Rate')
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(10,6))
churned_data = df[df['Churn'] == 1]
sns.boxplot(x='Churn',y='MonthlyCharges',data=churned_data,palette='rainbow',hue='Churn',legend=False)
plt.title('Churn Rate vs Monthly Charges')
plt.xlabel('Churn')
plt.ylabel('Monthly Charges')
plt.show()
```

```python
plt.figure(figsize=(10,6))
churned_data = df[df['Churn'] == 1]
sns.boxplot(x='Churn',y='TotalCharges',data=churned_data,palette='rainbow',hue='Churn',legend=False)
plt.title('Churn Rate vs Total Charges')
plt.xlabel('Churn')
plt.ylabel('Total Charges')
plt.show()
```

```python
tenure_churn = df.groupby('tenure')['Churn'].mean()

plt.figure(figsize=(10, 6))
plt.plot(tenure_churn.index, tenure_churn.values, marker='o', color='green', linestyle='-', linewidth=2)
plt.title('Churn Rate by Tenure')
plt.xlabel('Tenure (Months)')
plt.ylabel('Churn Rate')
plt.grid(True)
plt.show()
```

```python
contract_churn = df.groupby('InternetService')['Churn'].mean()  # Calculate churn rate for each contract type

plt.figure(figsize=(8, 6))
contract_churn.plot(kind='bar', color='orange')
plt.title('Churn Rate by Internet Service')
plt.xlabel('Contract')
plt.ylabel('Churn Rate')
plt.show()
```

```python
contract_churn = df.groupby('SeniorCitizen')['Churn'].mean()  # Calculate churn rate for each contract type

plt.figure(figsize=(8, 6))
contract_churn.plot(kind='bar', color='orange')
plt.title('Churn Rate by Type of Citizen')
plt.xlabel('SeniorCitizen-1')
plt.ylabel('Churn Rate')
plt.show()
```

```python
churned_data = df[df['Churn'] == 1]
plt.figure(figsize=(10, 6))
sns.histplot(churned_data['MonthlyCharges'], bins=20, kde=True, color='orange')
plt.title('Distribution of Monthly Charges for Churned Customers')
plt.xlabel('Monthly Charges')
plt.ylabel('Frequency')
plt.show()
```
